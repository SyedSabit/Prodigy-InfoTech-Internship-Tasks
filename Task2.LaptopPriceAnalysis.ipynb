## Relationship Between Manufacturere and prices

```python
plt.bar(df2["Manufacturere"],df2["Price"])
plt.xlabel('Manufacturere')
plt.ylabel('Price')
plt.title('Manufacturere vs Prices')
plt.grid(True)
plt.show()
```

## Relationship between RAM GB and prices

```python
plt.bar(df2["RAM GB"],df2["Price"])
plt.xlabel('RAM GB')
plt.ylabel('Price')
plt.title('RAM GB vs Prices')
plt.grid(True)
plt.show()
```
