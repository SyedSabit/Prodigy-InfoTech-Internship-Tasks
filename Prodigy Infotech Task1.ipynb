# Importing dependencies for analysis purpose

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### In this project we are going to analyze Pakistan Population from 1955 to present

```python
df = pd.read_csv(r"C:\Users\spark tech\Desktop\table-data.csv")
```

## Checking for missing values

```python
df.isnull().sum()
```

#### As we can see that their are no any missing values which is good for analysis purpose

```python
plt.bar(df["year"],df["population"])
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('pakistan Population Over Years')
plt.grid(True)
plt.show()
```

```python
plt.plot(df['year'], df['population'], marker='o')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Over Years')
plt.grid(True)
plt.show()
```

# Growth Rate

```python
plt.plot(df['year'], df['growthRate'], marker='o')
plt.xlabel('Year')
plt.ylabel('growthRate')
plt.title('PK GrowthRate Over Years')
plt.grid(True)
plt.show()
```

# PK population Rank in the world

```python
plt.plot(df['year'], df['populationRank'], marker='o')
plt.xlabel('Year')
plt.ylabel('populationRank')
plt.title('PK populationRank Over Years')
plt.grid(True)
plt.show()
```
